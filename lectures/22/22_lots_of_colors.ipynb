# RWantsHue

(https://github.com/hoesler/rwantshue)[https://github.com/hoesler/rwantshue]

(https://medialab.github.io/iwanthue/)[https://medialab.github.io/iwanthue/]

```r
library('rwantshue')
```

```r
scheme = iwanthue()
scheme$hex(10, color_space = hcl_presets$fluo)
colors = scheme$hex(15)
colors
```

```r
options(repr.plot.width = 7, repr.plot.height = 3.5)
p = ggplot(mpg, aes(x = displ, y = hwy, color = manufacturer))

p = p + geom_point()

p = p + scale_color_manual(values = colors)

p = p + guides(color = guide_legend(ncol=2))

p
```
